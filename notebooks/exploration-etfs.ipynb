```python
import pandas as pd
import plotly.express as px
from extraetf import etfs
```

```python
results1j = etfs.search(sort_by="return_1_year_real", ordering="desc", limit=5000)
```

```python
df_results = pd.DataFrame(results1j)
```

```python
df_results = df_results.dropna(subset=["costs"])
```

```python
df_results["costs_inv"] = 1.0 / df_results["costs"]*10
```

```python
px.scatter(df_results, x="sharpe_1", y="current_year", hover_data=["name", "isin"], color="volat_1", size="costs_inv")
```

```python
px.scatter(df_results, x="volat_1", y="year_1", hover_data=["name", "isin"], color="sharpe_1", size="costs_inv")
```

```python
px.scatter(df_results, x="year_1", y="drawdown_1", hover_data=["name", "isin"], color="sharpe_1", size="costs_inv")
```

```python
px.scatter(df_results, x="current_year", y="year_5", hover_data=["name", "isin"], color="sharpe_1", size="costs_inv")
```
